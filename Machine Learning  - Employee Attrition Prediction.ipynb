# Approach 2 - Multiple Linear Regression
In this approach, the second dataset (employees that have left) was imported in to the Dataframe and Multiple Linear Regression model was used to determine the types of employees that are leaving company X. 

1. The imported “employees that have left” dataset was encoded by converting the string categorical data in to float using “onehotencoder”
2. 
To avoid the dummy variable trap, index column 0 was sliced out of the dependent variable dataset.
3. The dataset was split in to 2 training and 2 test sets respectively using a test size of 0.28 (28%).
4. Multiple Linear Regression  algorithm was then fitted in to the training sets to predict the employee type that are prone  to leave the company.
5. The model was then optimized using “Backward Elimination Technique” so as to obtain the optimal predictor(s)/independent variable that have most significant effect on employee satisfaction level (dependent variable) which can determine whether an employee will stay in company X or leave.

## Approach

This problem was treated as a binary classification task because the objective was to predict whether an employee is likely to leave the company.

Three machine learning classification algorithms were implemented and evaluated:

* Decision Tree Classification
* Random Forest Classification
* Support Vector Machine (SVM) Classification

Each model was trained and evaluated using the same dataset, and their performance was compared using classification accuracy. The model with the highest accuracy was selected as the best-performing classifier.

To prepare the data, the dataset containing current employees was combined with the dataset of employees who had previously left the company. The resulting dataset contained both retained and former employees, enabling the machine learning models to learn patterns associated with employee attrition.

A new column named Employee Status was added to the combined dataset to serve as the target variable. Existing employees were assigned a value of 1, while employees who had left the company were assigned a value of 0. This column was used as the dependent variable (Y), while all remaining employee attributes were used as the independent variables (X).

Although the three classification algorithms differ in how they make predictions, they share several preprocessing steps. The categorical variables (Salary and Department) were encoded into numerical features using One-Hot Encoding so that they could be processed by the machine learning models. The dataset was then divided into training and testing sets using a 35% test split. Finally, feature scaling was applied to the Support Vector Machine (SVM) model to standardize the feature values, improving the model's performance and ensuring that variables with larger numerical values did not disproportionately influence the results.

```python
# Support Vector Machine (SVM) Classification
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
x = np.array(ohe.fit_transform(x), dtype=float)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=0)
print(x_train, x_test, y_train, y_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear', random_state=0)
classifier.fit(x_train, y_train)
```
